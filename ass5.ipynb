```python
import pandas as pd
import numpy as np
import scipy
```

```python
df = pd.read_csv("dat.csv")
```

```python
x = df['NIFTY'].values
y = df['NASDAQ'].values
```

```python
x
```

```python
x[1:10]
```

```python
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
print("Regression-" , r_value**2)
```

```python
average=0;
k=0;
for i in range(0,189):
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x[i:i+10], y[i:i+10])
    k = k+1
    average = average+r_value**2
    print("Regression-",i , r_value**2)
    
```
